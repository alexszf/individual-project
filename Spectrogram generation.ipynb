```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
y, sr = librosa.load("../recordings/baxter_room/AT2020 blu face.wav")
```

```python
D = np.abs(librosa.stft(y))
```

```python
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), y_axis='log', x_axis='time')
plt.title('Baxter Room: BT into AT2020 (facing)')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()
```

```python
# s10 mon face
y, sr = librosa.load("../Recordings/13th Nov/S10 blu away.wav")
D = np.abs(librosa.stft(y))
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), y_axis='log', x_axis='time')
plt.title('Baxter Room: BT into AT2020 (facing away)')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()
```
